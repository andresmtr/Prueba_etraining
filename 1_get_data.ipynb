### Encontrar sensor cercano a la tienda

```python
from geopy.distance import geodesic

# Lista para guardar sensor
tienda_sensor_match = []

for index, tienda in df[['tienda_id', 'latitud', 'longitud']].drop_duplicates().iterrows():
    tienda_coord = (tienda['latitud'], tienda['longitud'])
    min_dist = float('inf')
    sensor_mas_cercano = None

    for idx, sensor in ubicacion_sensores[['id', 'latitud', 'longitud']].iterrows():
        sensor_coord = (sensor['latitud'], sensor['longitud'])
        distancia = geodesic(tienda_coord, sensor_coord).km

        if distancia < min_dist:
            min_dist = distancia
            sensor_mas_cercano = sensor['id']
    
    tienda_sensor_match.append({
        'tienda_id': tienda['tienda_id'],
        'sensor_id_cercano': sensor_mas_cercano,
        'distancia_km': min_dist
    })


tienda_sensor_df = pd.DataFrame(tienda_sensor_match)
tienda_sensor_df.head()
```

```python
# Unir las ventas con su sensor más cercano
df = df.merge(tienda_sensor_df, on='tienda_id', how='left')
df
```

```python
# Verificar formato de fecha
sensor_data['fecha'] = sensor_data['fecha'].dt.floor('H')
df['fecha_venta'] = df['fecha_venta'].dt.floor('H')

# Merge
df = df.merge(
    sensor_data[['Sensor_id', 'fecha', 'valor']],
    left_on=['sensor_id_cercano', 'fecha_venta'],
    right_on=['Sensor_id', 'fecha'],
    how='left'
)

# Rellenar valores nulos de lluvia con 0
df['valor'] = df['valor'].fillna(0)

# Renombrar la columna 'valor' a 'lluvia_mm'
df = df.rename(columns={'valor': 'lluvia_mm'})

df.head()

```
